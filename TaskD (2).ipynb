How does the mean of the mean squared errors compare to that from Step B?

```python
#importing libraries

import keras
from keras.models import Sequential 
from keras.layers import Dense 
from sklearn.model_selection import train_test_split 
import pandas as pd 

#importing processing to perform normalization on predictors
from sklearn import preprocessing
```

```python
#loading dataset
data = pd.read_csv("concrete_data.csv")
data.head(10)
```

```python
#dividing dataset in predictors and target
predictors = data.iloc[:,:-1]
print(predictors)

#normalizing predictors
normalized_pred = preprocessing.normalize(predictors)
print(normalized_pred)

target = data.iloc[:,-1]
print(target)
```

```python
#building model 

model = Sequential()

n_cols = normalized_pred.shape[1]

#3 hidden layers each of 10 nodes and activation type relu
model.add(Dense(10, activation="relu", input_shape=(n_cols,)))
model.add(Dense(10, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
#printing mean squared error from 1-50 steps 
list = []
for step in range(50):
    #divide test data to be 30%
    X_train, X_test, y_train, y_test = train_test_split(normalized_pred, target, test_size=0.3)
    #train the model
    #changing epcohs from 50 to 100 for third task
    trained = model.fit(X_train, y_train, epochs=100, verbose=0, validation_data=(X_test, y_test))
    mean_squared_error = trained.history['val_loss'][-1]
    list.append(mean_squared_error)
    print('Step {} : mean squared error is {}'.format(step+1, mean_squared_error))
```

```python
print("The mean of the mean squared error of Task B was 111.18567626953126 and for Task D is: {}".format(np.mean(list)))
print("The standard deviation of the mean squared error is: {}".format(np.std(list)))
```
